Erozyon ve Genisleme

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

img = cv2.imread('text.png', 0)

kernel = np.ones((11,11), np.uint8)

img_erosion = cv2.erode(img, kernel, iterations = 1)
img_dilation = cv2.dilate(img, kernel, iterations = 1)



plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(img_erosion, 'gray')
plt.show()
plt.imshow(img_dilation, 'gray')
plt.show()
```

Acma

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

img = cv2.imread('text.png', 0)

kernel = np.ones((7,7), dtype = np.uint8)

whiteNoise = np.random.randint(0,2,size = img.shape[:2])
whiteNoise = whiteNoise * 255
noise_img = whiteNoise + img

openning = cv2.morphologyEx(noise_img.astype(np.float32),cv2.MORPH_OPEN,kernel)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(noise_img, 'gray')
plt.show()
plt.imshow(openning, 'gray')
plt.show()
```

Kapama

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

img = cv2.imread('text.png', 0)

kernel = np.ones((7,7), dtype = np.uint8)

blackNoise = np.random.randint(0,2,size = img.shape[:2])
blackNoise = blackNoise * -255
noise_img = blackNoise + img

closing = cv2.morphologyEx(noise_img.astype(np.float32),cv2.MORPH_CLOSE,kernel)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(noise_img, 'gray')
plt.show()
plt.imshow(closing, 'gray')
plt.show()
```
