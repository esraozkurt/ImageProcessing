```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('image.png')
kernel1 = np.ones((5,5), np.float32)/30
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)
res = np.hstack((image,img))

cv2.imshow('Original', image)
cv2.imshow('Kernel Blur', img)
cv2.waitKey()
cv2.destroyAllWindows()

plt.imshow(res)
plt.show()
```

```python
import cv2
import numpy as np
path = r'salad.jpg'
img = cv2.imread(path)
im1 = cv2.blur(img,(5,5))
im2 = cv2.boxFilter(img, -1, (2,2), normalize=True)

cv2.imshow('image', np.hstack((im1, im2)))
cv2.waitKey(0);
cv2.destroyAllWindows();
cv2.waitKey(1)

res = np.hstack((im1,im2))
plt.imshow(res)
plt.show()
```

```python
import cv2
import numpy as np
img = cv2.imread('xray.jpg')
median = cv2.medianBlur(img, 5)
compare = np.concatenate((img, median), axis=1)

cv2.imshow('img', compare)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(compare)
plt.show()
```
